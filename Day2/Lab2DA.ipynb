```python
df.loc[df["Name"] == "Khaled" , "Grade"] = 68
```

```python
df["Is_Excellent"] = np.where(df["Grade"] >= 90 , "Yes" , "No")
```

```python
print(df)
```

```python
df["Grade"].fillna(df["Grade"].mean())
```

```python
df["Score Date"].fillna(df["Score Date"].mode()[0], inplace=True)
```

```python
df["Final Grade"] = df["Grade"] + 5
print(df)
```

```python
bins = [0,69,79,89,100]
labels = ["Fail" , "Average" , "Good" , "Excellent"]
df["Grade Category"] = pd.cut(df["Grade"] , bins= bins , labels=labels ,include_lowest=True)
print(df)
```

```python
if 'Student ID' in df.columns:
    df.drop(columns=['Student ID'], inplace=True)
```

```python
df = df[df["Passed"] == True]
```

```python
print(df.sort_values('Grade' , ascending=False))
```

```python
print(df.sort_values(['City' , 'Grade'] ,ascending=[True , False] ))
```

```python
df['Rank'] = df['Grade'].rank(method='min', ascending=False)
print(df)
```

```python
print(df.nlargest(4 , 'Grade'))
```

```python
print(df.nsmallest(3 , "Age"))
```

```python
max_grade = df['Grade'].max()
print(df[df['Grade'] == max_grade][['Name', 'Grade']])
```
