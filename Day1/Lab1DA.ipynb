```python
data = np.array([150, 220, 90, 300, 180, 45])

nor = (data - data.min()) / (data.max() - data.min())

print(nor)
```

```python
data = np.array([150, 220, 90, 300, 180, 45])
z_score = (data - data.mean()) / data.std()
print(z_score)
```

```python
mat = np.array([
    [4, 8, 1, 12],
    [7, 3, 9, 5],
    [11, 2, 6, 10],
    [15, 0, 14, 13]
])
rmean = mat.mean(axis=1)
print(rmean)
```

```python
csum = mat.sum(axis=0)
print(csum)
```

```python
maxv = mat.max()
pos = np.unravel_index(np.argmax(mat), mat.shape)
print(maxv)
print(pos)
```
